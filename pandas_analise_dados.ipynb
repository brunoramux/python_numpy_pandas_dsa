# Group By e Reshape

```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais['CIDADE'].groupby(cidadesDigitais['UF']).count().sort_values(ascending=False)
# CONTAGEM AGRUPADO COM ORDENAÇÃO
```

```python
len(cidadesDigitais['UF'].unique())
# CONTAGEM DE ITENS UNICOS
```

```python
cidadesDigitais['STATUS'].unique()
```

```python
cidadesDigitais[cidadesDigitais['UF'] == 'SP'].count()
# CONTAGEM COM CONDICIONAL
```

```python
del cidadesDigitais['IBGE']
```

```python
# Aplicando Group By
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
type(cidadesUF_1)
```

```python
cidadesUF_1.groups
```

```python
# Número de grupos
len(cidadesUF_1.groups)
```

```python
ordenaUF = cidadesUF_1.size()
```

```python
# Ordena os grupos por ordem decrescente
ordenaUF.sort_values(ascending = False)
```

```python
cidadesUF_2 = cidadesDigitais.groupby(['UF','STATUS'])
# CONTAGEM AGRUPADA
```

```python
total = cidadesUF_2.size()
total.sort_values(ascending = False)
total
```

```python
cidadesDigitais.head()
```

```python
cidadesUF_3 = cidadesDigitais.set_index('STATUS')
# SET INDEX É UTILIZADO PARA FAZER PIVOTAGEM NA TABELA, TRANSFORMANDO LINHAS EM COLUNAS OU COLUNAS EM LINHAS
# DESSA FORMA, PODEMOS AGRUPAR E SUMARIZAR DADOS MAIS FACILMENTE USANDO O PANDAS E O NUMPY
```

```python
# Aplicando Group By
cidadesUF_3 = cidadesUF_3.groupby(level = 'STATUS')
```

```python
cidadesUF_3.head()
```

```python
cidadesUF_4 = cidadesDigitais.set_index(['STATUS', 'UF'])
```

```python
cidadesUF_4
```

```python
cidadesUF_4 = cidadesDigitais.set_index(['UF', 'STATUS'])
```

```python
cidadesUF_4 = cidadesUF_4.groupby(level = ['UF', 'STATUS'])
```

```python
cidadesUF_4.sum()
```

```python
cidadesUF_4.aggregate(np.sum)
# USAMOS A FUNCAO AGGREGATE PARA SUMARIZAR DADOS DE ACORDO COM O INDICES SETADOS
```

```python
cidadesUF_4.aggregate([np.sum, np.mean, np.size])
# USANDO O NUMPY, PODEMOS FAZER VÁRIAS SUMARIZAÇÕES DE UMA SÓ VEZ
```

```python
cidadesUF_5 = cidadesDigitais.set_index([ 'STATUS', 'CIDADE', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO', 'POPULAÇÃO', 'PONTOS_ATENDIDOS'])
```

```python
cidadesUF_5 = cidadesUF_5.groupby(level = ['STATUS', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO'])
```
